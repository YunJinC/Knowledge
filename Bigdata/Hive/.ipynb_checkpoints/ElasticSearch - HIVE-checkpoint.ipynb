### ElasticSearch Index 생성
- Hive 테이블을 ElasticSearch와 연동 하기 위한 Index 생성
```SHELL
curl -XPUT http://IPADDRESS:PORT/undata/?pretty -d'
{
 "settings" : {
  "number_of_shards" : 1
 },
 "mappings" : {
  "rel" : {
   "properties": {
    "period":           {"type":"string"},
    "trade_flow_code":  {"type":"string"},
    "reporter_code":    {"type":"string"},
    "partner_code":     {"type":"string"},
    "commodity_code":   {"type":"string"},
    "trade_value_us":   {"type":"string"}
   }
  }
 }
}'
```

### Hive 테이블 생성
- Hive 테이블을 ElasticSearch와 연동 하기 위한 hive 테이블 생성(Zeppelin or Hive Shell)
```SQL
DROP TABLE TABLE_NAME;
CREATE EXTERNAL TABLE 
    TABLE_NAME
    (
        period               string, 
        trade_flow_code      string,        
        reporter_code        string,        
        partner_code         string,        
        commodity_code       string,        
        trade_value_us       float
    )
STORED BY 'org.elasticsearch.hadoop.hive.EsStorageHandler'
TBLPROPERTIES('es.resource' = 'undata/rel', 'es.nodes' = IPADDRESS, 'es.index.auto.create' = 'false');
```
- STORED BY : 'org.elasticsearch.hadoop.hive.EsStorageHandler’
- es.resource = index/type, es.nodes = 저장노드

### Hive 테이블에 적재
```sql
insert OVERWRITE table kotra_ws.elastic_tb_uncd_year_2016
select period, trade_flow_code, reporter_code, partner_code, commodity_code, trade_value_us from kotra_ws.tb_uncd_year_2016;
```
